```python
import altair as alt
import pandas as pd


data_path = 'data/AUM_V4_Activity_2018-06-21_17-16-27.csv'
```

```python
df = pd.read_csv(data_path, parse_dates=[['Date', 'Time']])
df = df[:-3]
df = df.set_index('Date_Time')
# index = pd.DatetimeIndex(df.Date_Time)
# df = pd.Series(df['App name', 'Duration'], index=index)
df.head(10)
```

```python
df.dtypes
```

## Convert Duration to TimeDelta

```python
df['Duration'] = pd.to_timedelta(df.Duration)
df.index = pd.to_datetime(df.index)
df.head()
```

```python
df.index = df.index.to_series().dt.to_period('H')
```

```python
df.sample(10)
```

```python
tinder = df.loc[df['App name'] == "Tinder"]
tinder.head(10)
```

```python
import numpy as np


# grouped = tinder.groupby("Date_Time")
# grouped['Duration'].aggregate(np.sum).head(10)

# tinder = tinder.groupby(['Date_Time','App name'])['Duration'].sum()
tinder = tinder.resample('H')['Duration'].sum()
tinder = tinder.to_frame()
```

```python
chart = alt.Chart(tinder.to_frame).mark_bar().encode(
    x='Date_Time:o',
    y='Duration:o'
)

chart
```
